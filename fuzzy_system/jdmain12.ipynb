```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from pprint import pprint
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# visualization


%matplotlib inline


# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
# fix random seed for reproducibility
numpy.random.seed(7)
# load pima indians dataset
filename1 = pd.read_csv('heart-statlog.csv')
# split into input (X) and output (Y) variables

X_train=filename1.iloc[0:650,0:12]
Y_train=filename1.iloc[0:650,12:13]

X_test=filename1.iloc[651:872,0:12]
Y_test=filename1.iloc[651:872,12:13]
# create model
model = Sequential()
model.add(Dense(25, input_dim=12, activation='tanh'))
model.add(Dense(15, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
Y_train
```

```python
model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy'])
# Fit the model
```

```python
model.fit(X_train, Y_train,validation_data=(X_test, Y_test), nb_epoch=200, batch_size=64)
# evaluate the model
```
